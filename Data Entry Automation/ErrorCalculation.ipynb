```python
import pandas as pd
import numpy as np
i = 0
while i < 11:
 with open("excel1ipi%i.csv" %i, 'r') as temp_f:
    # get No of columns in each line
    col_count = [ len(l.split(",")) for l in temp_f.readlines() ]

### Generate column names  (names will be 0, 1, 2, ..., maximum columns - 1)
 column_names = [i for i in range(max(col_count))] 




 df2 = pd.read_csv('excel1ipi%i.csv'%i,header=None,names=column_names)
 df1 = pd.read_csv('P1%i.csv'%i,header=None,names=column_names)
 df1 = df1.replace(np.nan, '', regex=True)
 df2 = df2.replace(np.nan, '', regex=True)


 df1.equals(df2)
 comparison_values = df1.values == df2.values
 print(comparison_values)
 rows,cols=np.where(comparison_values==False)

 error = 0
 for item in zip(rows,cols):
    df2.iloc[item[0], item[1]] = '{} error'.format(df2.iloc[item[0], item[1]])
    error = error+1
 def red(val):
  
    if "error" in val:
        color = 'red'
    else:
        color = 'black'
    return 'color: %s' % color

 df3 =df2.style.applymap(red)
 s = 0
 Total_Values = 0
 for x in df2.values:
    for y in x:
        if y == "":
            s = s+1
        else:
            Total_Values = Total_Values+1
 Percentage_error = (error/Total_Values)*100
 print("Total values = {}".format(Total_Values))

 print("Total Incorrect values = {}".format(error))
 print("Percentage Error = {}".format(Percentage_error))

 print(Percentage_error)

 df3.to_excel("finalerror2%i.xlsx" %i,index = False,header = False)
 df1.to_excel("correct2%i.xlsx" %i,index = False,header = False)
 i = i+1
```
